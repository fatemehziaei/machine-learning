sin function and noise

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 4 * np.pi, 100)
y=np.sin(x)
y_noise=y+0.1* np.random.randn(*x.shape)
plt.plot(x,y,'r-')
plt.scatter(x,y_noise,c='b',marker="o")
plt.show()
```
